## 1. The mortality rate (Crude death and standardize death)

The crude death is the straightforward way to measure the death rate, the formula (per 1000 people) is

$CDR = 1000\times(\frac{N_{deaths}}{N_{population}})$

$CDR =$ Crude death rates
<br>
$N_{deaths} =$ Number of deaths occurred
<br>
$N_{population} =$ Number of population observed

On the other hand, the standardize rate is the rate that is weighted on the risk of death because the older one get the higher chance of death is increasing. Thus, the formula (per 1000 people) is

$SDR = \sum_{age} \frac{m_{age}p_{age}}{p_{age}}$

$SDR =$ Standardized death rates
<br>
$m_{age} =$ The death rate of the specific $age$
<br>
$p_{age} =$ The number of population observed of the specific $age$

"Glossary:Standardised death rate (SDR)," Eurostat, accessed December 2, 2022, https://ec.europa.eu/eurostat/statistics-explained/index.php?title=Glossary:Standardised_death_rate_(SDR)

The means the crude death rate count the number of death, while standardized death rate is counting by weighting the expected cases higher than the unexpected cases. Thus, the higher the standardized rate is, the more expected death happened (which means less accidents or other kind of immediate deaths.)

### Data processing

Create a new column named "Type of death" to store the types of death and then use that to visualize data between crude death rate and standardize death rate.

```python
# Do some more data processing by create a new variable which is type of death so that the data can be visualized by separated the death types.
# we want to plot the rate in both crude death rate and standardize death rate, so we do a few changes in the dataset.
df_2 = df.melt(id_vars=['Yr'], value_vars=['Crude_death_r', 'Stand_death_r'], var_name='Type of death', value_name='Rate (per 1000 population)')
df_2
```

### Data visualization and Data modelling

Create the area graph which represents the real value of the death rates, and the line graph which represent the trends of the data (linear regression trends)

```python
area = alt.Chart(df_2).mark_area(opacity=0.3).encode(
    x = 'Yr',
    y = alt.Y('Rate (per 1000 population)', stack=None),
    color = 'Type of death'
)

base = alt.Chart(df).mark_circle(opacity=0.7).transform_fold(
    fold = ['Crude_death_r', 'Stand_death_r'],
    as_  = ['Type of death', 'Rate (per 1000 population)']
).encode(
    alt.X('Yr'),
    alt.Y('Rate (per 1000 population):Q'),
    alt.Color('Type of death:N')
)

area + base.transform_regression('Yr', 'Rate (per 1000 population)', groupby=['Type of death']).mark_line(size=3)
```

### Converting the data into equation/Observing the trends

The trend from the graph above seems to be the linear regession graph because the linear graph fit the trends very well. The data will be converted into the equation to make some predictions.

```python
# from the graph, it seems like the trend is linear
trendline_crude = np.polyfit(df['Yr'], df['Crude_death_r'], 1)
trendline_stand = np.polyfit(df['Yr'], df['Stand_death_r'], 1)

print('The linear equation of the crude death rate is Crude_death_r = Yr *', trendline_crude[0], '%+.14f' % (trendline_crude[1]))
print('The linear equation of the standardiize death rate is Stand_death_r = Yr *', trendline_stand[0], '%+.14f' % (trendline_stand[1]))
```

### Data interpretation

After observing the trends and getting the linear equation, use the equation to predict the trend. In this project, the values will be predicted up to year 2050.

```python
# predict the year up to 2050

for i in range(2017, 2051):
    df_pred = pd.DataFrame({'Yr': [i],
                        'Crude_death_r': [i*trendline_crude[0] + trendline_crude[1]],
                        'Stand_death_r': [i*trendline_stand[0] + trendline_stand[1]]})
    df = pd.concat([df, df_pred], ignore_index=True)
df
```

After got the values in dataset, the visualization is the same as the above one

```python
df_2 = df.melt(id_vars=['Yr'], value_vars=['Crude_death_r', 'Stand_death_r'], var_name='Type of death', value_name='Rate (per 1000 population)')
df_2

area = alt.Chart(df_2).mark_area(opacity=0.3).encode(
    x = 'Yr',
    y = alt.Y('Rate (per 1000 population)', stack=None),
    color = 'Type of death'
)

base = alt.Chart(df).mark_circle(opacity=0.7).transform_fold(
    fold = ['Crude_death_r', 'Stand_death_r'],
    as_  = ['Type of death', 'Rate (per 1000 population)']
).encode(
    alt.X('Yr'),
    alt.Y('Rate (per 1000 population):Q'),
    alt.Color('Type of death:N')
)

area + base.transform_regression('Yr', 'Rate (per 1000 population)', groupby=['Type of death']).mark_line(size=3)
```

### Analyze the difference of the data

As the graph suggested, there are 19 unique causes (excluding the category "other cases") of death during 1986-2000 and 2006-2016. Therefore, the author wants to find the unique cases that exclusively have in 1986-2000 period and 2006-2016 period, and the cases that both periods have. 

```python
# the unique causes of death are not the same (in terms of amount and context)

def count_unique(array):
    for i in range(len(array)):
        for j in range(len(array)-1, i+1, -1):
            if array[i] == array[j] and array[i] != 0 and array[j] != 0:
                array[j] = 0
    
    count = len(array)            
    for i in range(len(array)):
        if array[i] == 0:
            count -= 1
    return count

df1_copy = df1.copy()

print('The number unique cause of death from 1986-2000 is', count_unique(df1_copy['COD']), ', which are')
for i in df1_copy['COD']:
    if i != 0:
        print(i)
```

```python
df2_copy = df2.copy()
print('The number unique cause of death from 2006-2016 is', count_unique(df2_copy['COD']), ', which are')
for i in df2_copy['COD']:
    if i != 0:
        print(i)
```

In the author opinion, the trends of both countries share the similarity trends. The graphs are steep in the 1986-2006, and then slightly grow afterwards. Which means the trendlines are logarithmic function. Then fill all the Germany's and U.S.A.'s missing values.

```python
# For me, it seems both of the countries have trendline with logarithmic function
# starts with Germany

df_ger[['Yr', 'Exp_of_life_birth']] = df_ger[['Yr', 'Exp_of_life_birth']].apply(pd.to_numeric)

# male
df_ger_m = df_ger[df_ger['Sex'] == 'Male']
curve_m = np.polyfit(np.log(df_ger_m['Yr']), df_ger_m['Exp_of_life_birth'], 1)

# the missing value is Yr = 1986 (col 6), 2015 (col 222), 2016 (col 246)
df.iat[6, 3] = np.log(df.iat[6, 0])*curve_m[0] + curve_m[1]
df.iat[222, 3] = np.log(df.iat[222, 0])*curve_m[0] + curve_m[1] 
df.iat[246, 3] = np.log(df.iat[246, 0])*curve_m[0] + curve_m[1]
df[df['Economy'] == 'Germany'][df['Sex'] == 'Male']
```

```python
# female
df_ger_f = df_ger[df_ger['Sex'] == 'Female']
curve_f = np.polyfit(np.log(df_ger_f['Yr']), df_ger_f['Exp_of_life_birth'], 1)

# the missing value is Yr = 1986 (col 18), 2015 (col 234), 2016 (col 258)
df.iat[18, 3] = np.log(df.iat[18, 0])*curve_f[0] + curve_f[1]
df.iat[234, 3] = np.log(df.iat[234, 0])*curve_f[0] + curve_f[1] 
df.iat[258, 3] = np.log(df.iat[258, 0])*curve_f[0] + curve_f[1]
df[df['Economy'] == 'Germany'][df['Sex'] == 'Female']
```

```python
# U.S.A.
df_usa[['Yr', 'Exp_of_life_birth']] = df_usa[['Yr', 'Exp_of_life_birth']].apply(pd.to_numeric)

# male
df_usa_m = df_usa[df_usa['Sex'] == 'Male']
curve_um = np.polyfit(np.log(df_usa_m['Yr']), df_usa_m['Exp_of_life_birth'], 1)

# the missing value is Yr = 2016 (Col 251)
df.iat[251, 3] = np.log(df.iat[251, 0])*curve_um[0] + curve_um[1]

# female
df_usa_f = df_usa[df_usa['Sex'] == 'Female']
curve_uf = np.polyfit(np.log(df_usa_f['Yr']), df_usa_f['Exp_of_life_birth'], 1)

# the missing value is Yr = 2016 (Col 263)
df.iat[263, 3] = np.log(df.iat[263, 0])*curve_uf[0] + curve_uf[1]
df[df['Economy'] == 'U.S.A.']
```
